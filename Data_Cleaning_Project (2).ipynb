```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.countplot(data=df,x='Machine_Speed',hue='Machine_Speed',palette='dark')
plt.ylabel('Count')
plt.yticks(range(0,10))
plt.grid(axis='y')
plt.title('Washing Maching Machine Speed counts')
plt.show()
```

```python
df.iloc[:,7].plot(kind='hist',bins= 100)
plt.title('Frequency of Machine Speed')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Offer_Price'], bins=15, kde=True, color='lightgreen')
plt.title('Price Distribution')
plt.show()
```

```python
df['Offer_Price'].value_counts().plot(kind='bar')
plt.xlabel('Price')
plt.yticks(range(0,))
plt.title('price Frequency ')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(df['Offer_Price'])
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(y='Brand_Name', data=df, order=df['Brand_Name'].value_counts().index)
plt.title('Brand Frequency')
plt.show()
```

```python
sns.histplot(data=df,x=df['Actual_Price'],label='Actual Price',kde=True)
sns.histplot(data=df,x=df['Offer_Price'],label='Price After Discount',kde=True)
plt.title('Frequency of Actual Price')
plt.xlabel('Actual Price')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
sns.histplot(data=df,x=df['Offer'],label='Offer',hue='Brand_Name',bins=50)
plt.xticks(rotation=90)
plt.xticks(range(0,80,15))
plt.title('Frequency of Offer')
plt.xlabel('Offer')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot Pie Chart
offer_counts = df['Brand_Name'].value_counts()
max_value = offer_counts.max()
explode = [0.1 if i == max_value else 0 for i in offer_counts]
df['Brand_Name'].value_counts().plot(kind='pie', autopct='%1.1f%%',explode=explode, startangle=60, colormap='viridis')
plt.title('Offer Distribution (%)')
```

```python
sns.kdeplot(df['Offer_Price'])
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.barplot(y=df['Brand_Name'],x=df['Offer_Price'])
plt.xticks(rotation=30)
plt.title('Brand and Price')
plt.ylabel('Brand Name')
plt.xlabel('Price')
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=[np.number])
plt.figure(figsize=(6,4))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pairplot of selected columns
sns.pairplot(df[['Rating', 'Offer_Price', 'Actual_Price', 'Weight_Capacity (kg)', 'Brand_Name']], hue='Brand_Name')
plt.suptitle('Multivariate - Pairplot', y=1.02)
plt.show()
```
