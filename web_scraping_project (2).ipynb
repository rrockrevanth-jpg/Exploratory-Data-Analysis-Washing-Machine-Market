```python
import re 
import pandas as pd 
import numpy as np
```

```python
Washing_machine_name=[]
Weight=[]
rating_1=[]
Ratings=[]
Reviews=[]
price=[]
Speed=[]
Actual=[]
Offer=[]
Comprehensive=[]
Warr_wash_motor=[]
Spin_motor=[]
Dit_motor=[]

for i in soup.find_all("div",class_="tUxRFH"): # main container
    
    # price from main container 
    p=i.find("div",class_="Nx9bqj _4b5DiR")
    if p:
        price.append(p.text)
    else:
        price.append(np.nan)
    
    # rating_1 from main container 
    r1=i.find("div",class_="XQDdHH")
    if r1:
        rating_1.append(r1.text)
    else:
        rating_1.append(np.nan)
    
    # actual price 
    actual=i.find("div",class_="yRaY8j ZYYwLA")
    if actual:
        Actual.append(actual.text)
    else:
        Actual.append(np.nan)  
    
    
    ############
    fo=i.find("div",class_="col col-5-12 BfVC2z")
    if fo:
        #print(fo.text)
        offer=re.findall(r"^\₹\d.+(\d\d\%)\s",fo.text)
        if offer:
            Offer.append(offer[0])
        else:
            Offer.append(np.nan)
            
    
    
    foinn=i.find("span",class_="Wphh3N")
    if foinn:
        #print(foinn.text)
        rating=re.findall(r"(\d.+|\d)\sRatings",foinn.text)
        if rating:
            #print(rating)
            Ratings.append(rating[0])
        else:
            Ratings.append(np.nan)

         # washing machine rating
        reviews=re.findall(r"\&\s(\d.+|\d)\sReviews",foinn.text)
        if reviews:
            #print(reviwes)
            Reviews.append(reviews[0])
        else:
            Reviews.append(np.nan)
    
    foin=i.find("div",class_="KzDlHZ")
    if foin:
        #print(foin.text)
        # washing machine information name
        washing=re.findall(r"(^\w.+)\s\d\.?\d?\skg",foin.text)
        if washing:
            #print(washing)
            Washing_machine_name.append(washing[0])
        else:
            Washing_machine_name.append(np.nan)

        # washing machine weight
        weight=re.findall(r"(\d\.?\d?)\skg",foin.text)
        if weight:
            #print(weight)
            Weight.append(weight[0])
        else:
            Weight.append(np.nan)
    
    
    info=i.find("ul",class_="G4BRas") 
    if info:
        
        #print(info.text)
        # washing machine information 1    
        speed=re.findall(r"(\d+|\d.+)\s\w+\sMax",info.text)
        if speed:
            #print(speed)
            Speed.append(speed[0])
        else:
            Speed.append(np.nan)

        
        
        # Comprehensive Warranty
        comprehensive=re.findall(r"[Heater|Rating]?(\d+)\s?[yY]ear[sS]?\s\w+",info.text)
        if comprehensive:
            Comprehensive.append(comprehensive[0])
        else:
            Comprehensive.append(np.nan)
        
        
        # Warranty on motor
        warr_wash_motor=re.findall(r"(\d+)\s?\w+?\s?\w+?\s?\w+\s?Motor",info.text)
        if warr_wash_motor:
            Warr_wash_motor.append(warr_wash_motor[0])
        else:
            Warr_wash_motor.append(np.nan)
        
        # Warranty on spin motor
        spin_motor=re.findall(r"\s?(\d+)\s?\w+?\s?\w+?\sSpin\smotor",info.text)
        if spin_motor:
            Spin_motor.append(spin_motor[0])
        else:
            Spin_motor.append(np.nan)

        # DIT motor
        dit_motor=re.findall(r"\,?\s?(\d+)\s?\w+?\s?\w+?\s?\w+?\s?\w+?\s?DIT\s[Mm]otor",info.text)
        if dit_motor:
            Dit_motor.append(dit_motor[0])
        else:
            Dit_motor.append(np.nan)







doc={"Brand_Name":Washing_machine_name,"Weight_Capacity (kg)":Weight,"Rating":rating_1,"Total_no_people_ratings":Ratings,"Total_no_people_reviews":Reviews,
     "Actual_Price":Actual,"Offer_Price":price,"Offer":Offer,"Machine_Speed":Speed,"Comprehensive_Warranty (Years)":Comprehensive,
     "Wash_motor_Warranty (Years)":Warr_wash_motor,"Spin_motor_Warranty (Years)":Spin_motor,"DIT_motor_Warranty (Years)":Dit_motor}

df=pd.DataFrame(doc)
df
        
```

```python
Washing_machine_name=[]
Weight=[]
rating_1=[]
Ratings=[]
Reviews=[]
Actual=[]
price=[]
Speed=[]
Offer=[]
Comprehensive=[]
Warr_wash_motor=[]
Spin_motor=[]
Dit_motor=[]

for page in range(1,40):
    url=f"https://www.flipkart.com/search?q=washing%20machine&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as={page}"
    req=requests.get(url,headers=request_header)
    soup=bs(req.text,"html.parser")
    for i in soup.find_all("div",class_="tUxRFH"): 
        # main container
        # price from main container 
        p=i.find("div",class_="Nx9bqj _4b5DiR")
        if p:
            price.append(p.text)
        else:
            price.append(np.nan)
        
        # rating_1 from main container 
        r1=i.find("div",class_="XQDdHH")
        if r1:
            rating_1.append(r1.text)
        else:
            rating_1.append(np.nan)
        # actual price 
        actual=i.find("div",class_="yRaY8j ZYYwLA")
        if actual:
            Actual.append(actual.text)
        else:
            Actual.append(np.nan)  

        ############
        fo=i.find("div",class_="col col-5-12 BfVC2z")
        if fo:
            #print(fo.text)
            offer=re.findall(r"^\₹\d.+(\d\d\%)\s",fo.text)
            if offer:
                Offer.append(offer[0])
            else:
                Offer.append(np.nan)
            
        foinn=i.find("span",class_="Wphh3N")
        if foinn:
            #print(foinn.text)
            rating=re.findall(r"(\d.+|\d)\sRatings",foinn.text)
            if rating:
                #print(rating)
                Ratings.append(rating[0])
            else:
                Ratings.append(np.nan)
    
             # washing machine rating
            reviews=re.findall(r"\&\s(\d.+|\d)\sReviews",foinn.text)
            if reviews:
                #print(reviwes)
                Reviews.append(reviews[0])
            else:
                Reviews.append(np.nan)
        
        foin=i.find("div",class_="KzDlHZ")
        if foin:
            # washing machine information name
            washing=re.findall(r"(^\w.+)\s\d\.?\d?\skg",foin.text)
            if washing:
                #print(washing)
                Washing_machine_name.append(washing[0])
            else:
                Washing_machine_name.append(np.nan)
    
            # washing machine weight
            weight=re.findall(r"(\d\.?\d?)\skg",foin.text)
            if weight:
                #print(weight)
                Weight.append(weight[0])
            else:
                Weight.append(np.nan)
        
        
        info=i.find("ul",class_="G4BRas") 
        if info:
            
            #print(info.text)
            # washing machine information 1    
            speed=re.findall(r"(\d+|\d.+)\s\w+\sMax",info.text)
            if speed:
                #print(speed)
                Speed.append(speed[0])
            else:
                Speed.append(np.nan)
    
            

             # Comprehensive Warranty
            comprehensive=re.findall(r"[Heater|Rating]?(\d+)\s?[yY]ear[sS]?\s\w+",info.text)
            if comprehensive:
                Comprehensive.append(comprehensive[0])
            else:
                Comprehensive.append(np.nan)

            
            # Warranty on motor
            warr_wash_motor=re.findall(r"(\d+)\s?\w+?\s?\w+?\s?\w+\s?Motor",info.text)
            if warr_wash_motor:
                Warr_wash_motor.append(warr_wash_motor[0])
            else:
                Warr_wash_motor.append(np.nan)

            # Warranty on spin motor
            spin_motor=re.findall(r"\s?(\d+)\s?\w+?\s?\w+?\sSpin\smotor",info.text)
            if spin_motor:
                Spin_motor.append(spin_motor[0])
            else:
                Spin_motor.append(np.nan)
            # DIT motor
            dit_motor=re.findall(r"\,?\s?(\d+)\s?\w+?\s?\w+?\s?\w+?\s?\w+?\s?DIT\s[Mm]otor",info.text)
            if dit_motor:
                Dit_motor.append(dit_motor[0])
            else:
                Dit_motor.append(np.nan)


    
        


doc={"Brand_Name":Washing_machine_name,"Weight_Capacity (kg)":Weight,"Rating":rating_1,"Total_no_people_ratings":Ratings,"Total_no_people_reviews":Reviews,
     "Actual_Price":Actual,"Offer_Price":price,"Offer":Offer,"Machine_Speed":Speed,"Comprehensive_Warranty (Years)":Comprehensive,
     "Wash_motor_Warranty (Years)":Warr_wash_motor,"Spin_motor_Warranty (Years)":Spin_motor,"DIT_motor_Warranty (Years)":Dit_motor}

df=pd.DataFrame(doc)
df
        
```
